```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
import nltk
import numpy as np
import re
from nltk.corpus import stopwords
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
path="C:/Users/owner/Downloads/sms.csv"
df=pd.read_csv(path,header=0,encoding='latin-1')
df
```

```python
df=df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
df
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df
```

```python
x=df.iloc[:,0]
y=df.iloc[:,1]
print(x)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(X_train, X_test, y_train, y_test)
```

```python
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
print(X_train_tfidf,X_test_tfidf)
```

```python
svm_model = SVC(kernel='linear')  # 'linear', 'rbf', 'poly'
svm_model.fit(X_train_tfidf, y_train)
print(svm_model)
predictions = svm_model.predict(X_test_tfidf)
print(predictions)
```

```python
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:\n', classification_report(y_test, predictions))
```

```python
svm_model1 = SVC(kernel='rbf')  # 'linear', 'rbf', 'poly'
svm_model1.fit(X_train_tfidf, y_train)
svm_model1
pred1 = svm_model1.predict(X_test_tfidf)
print(pred1)
```

```python
accuracy = accuracy_score(y_test, pred1)
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:\n', classification_report(y_test, pred1))
```
